```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data_filepath = "/content/El Pilar_Samples_20250310.csv"

data = pd.read_csv(data_filepath)

data.head(10)
```

```python
data.describe()
```

```python
filtered_df = data[(data['Sample_Mass'] > 0) & (data['Sample_Mass'] < 673839.000000)]
filtered_df.head()
```

```python
filtered_df = filtered_df.copy()  # Ensure it's a new DataFrame
filtered_df['RecvdWt_kg_WEI-21'] = filtered_df['RecvdWt_kg_WEI-21'].astype(float)
```

```python
filtered_df['RecvdWt_kg_WEI-21'] = filtered_df['RecvdWt_kg_WEI-21'].fillna(filtered_df['RecvdWt_kg_WEI-21'].mean())
```

```python
filtered_df.describe()
```

```python
filtered_df['Relative_Difference'] = abs(
    (filtered_df['Sample_Mass'] - filtered_df['RecvdWt_kg_WEI-21']) /
    (0.5 * (filtered_df['Sample_Mass'] + filtered_df['RecvdWt_kg_WEI-21']))
) * 100

filtered_df.head()
```

```python
# Identify groups where the Collar column changes
filtered_df['Group'] = (filtered_df['Collar'] != filtered_df['Collar'].shift()).cumsum()

# Split the DataFrame into a dictionary of DataFrames
groups = {key: group.drop(columns='Group') for key, group in filtered_df.groupby('Group')}
```

```python
len(groups)
```

```python
fig = go.Figure()

# Add traces for each group
for i, (key, group) in enumerate(groups.items()):
    fig.add_trace(go.Scatter(
        x=group['Sample_Mass'],
        y=group['RecvdWt_kg_WEI-21'],
        mode='markers',
        name=f'Group {key}',
        visible=(i == 0)  # Show only the first group initially
    ))

# Create dropdown buttons
dropdown_buttons = []
for i, (key, _) in enumerate(groups.items()):
    visibility = [False] * len(groups)
    visibility[i] = True  # Show only the current group
    dropdown_buttons.append({
        'args': [{'visible': visibility}],
        'label': f'Log {groups[key].iat[0, 0]}',
        'method': 'update'
    })

# Update layout with dropdown
fig.update_layout(
    updatemenus=[{
        'buttons': dropdown_buttons,
        'direction': 'down',
        'showactive': True,
        'pad': {'r': 10},
    }],
    title='Scatter Plot',
    xaxis_title='Sample_Mass',
    yaxis_title='RecvdWt_kg_WEI-21',
)

fig.show()
```
